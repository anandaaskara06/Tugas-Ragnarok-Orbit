```python
#parent
class Karyawan(object) :
    def __init__(self, nama, umur, lamaKerja) :
        self.nama = nama
        self.umur = umur
        self.__lamaKerja = lamaKerja
        
    def info(self) :
        print(f"\nnama: {self.nama} \numur: {self.umur} \nlamaKerja: {self.__lamaKerja} tahun")

    #getter
    def getLamaKerja(self):
        return self.__lamaKerja
    
    #setter
    def tambahLamaKerja(self, tahun) :
        self.__lamaKerja += tahun
        
    #method parent
    def gajian(self) :
        print(self.nama, "gajian")
        
#child inheritence ke parents
class Marketing(Karyawan) :
    def __init__(self,  nama, umur, lamaKerja, gaji, posisi):
        
        #milik parents
        super().__init__(nama, umur, lamaKerja)
        
        #milik child
        self.gaji = gaji
        self.posisi = posisi

    def infoMark(self):
        
        #memanggil method parents
        super().info()
        #menjalankan method child
        print(f"gaji: {self.gaji} \nposisi: {self.posisi}")

mark1 = Marketing ("Raymond", 27, 7, "7 juta", "pegawai tetap")
mark1.infoMark()
mark1.tambahLamaKerja(1)
mark1.infoMark()
```

ABSCTRATION

```python
class Absc(ABC):
    @abstractmethod
    def penugasan(self):
        pass
    
#class karyawan inheritence ke kelas abstract    
class Karyawan(Absc) :
    def __init__(self, nama, umur, lamaKerja) :
        
        #biasa
        self.nama = nama
        self.umur = umur
        
        #private
        self.__lamaKerja = lamaKerja
        
    def info(self) :
        print(f"\nnama: {self.nama} \numur: {self.umur} \nlamaKerja: {self.__lamaKerja} tahun")

    #getter
    def getLamaKerja(self):
        return self.__lamaKerja
    
    #setter
    def tambahLamaKerja(self, tahun) :
        self.__lamaKerja += tahun
        
    #method biasa
    def gajian(self) :
        print(self.nama, "gajian")
        
    #abstract method dari class Absc
    def penugasan(self, tempat) :
        print(self.nama,"ditugaskan ke",tempat)
        
#class marketing inheritence ke class karyawan dan absc        
class Marketing(Karyawan, Absc) :
    def __init__(self,  nama, umur, lamaKerja, gaji, posisi):
        
        #turunan parents
        super().__init__(nama, umur, lamaKerja)
        
        #milik child
        self.gaji = gaji
        self.posisi = posisi

    def infoMark(self):
        #memanggil method parents
        super().info()
        #method child
        print(f"gaji: {self.gaji} \nposisi: {self.posisi}")
    
    #akses abstract method dari super
    def penugasanMark(self,tempat) :
        super().penugasan(tempat)
        
    #akses abstract method langsung ke kelas Absc
    def penugasan(self) :
        print(self.nama,"sedang bertugas")

mark1 = Marketing ("Raymond", 27, 7, "7 juta", "pegawai tetap")
mark1.infoMark()
mark1.tambahLamaKerja(1)
mark1.infoMark()
mark1.penugasanMark("jakarta")
mark1.penugasan()
```

OOP 2 Kasus Buah-buahan

ENCAPSULATION, INHERITENCE, ABSCTRACTION DAN POLIMORPHISM DIGABUNG

```python
#Abstraction
class Abstract2(ABC):
    @abstractmethod
    def ditanam() :
        pass
    
#class karyawan inheritence ke kelas abstract    
class Buah(Abstract2) :
    def __init__(self, warna, biji) :

        self.warna = warna
        self.biji = biji    
        
    def info(self) :
        print(f"\nwarna: {self.warna} \numur: {self.biji}")
        
    #method biasa
    def dipetik(self) :
        print("buah matang berwarna", self.warna, "siap dipetik")
        
    #abstract method dari class Absc
    def ditanam(self) :
        print("buah ditanam hingga berwarna", self.warna)
        
#class melon inheritence ke class buah        
class Melon(Buah) :
    def __init__(self,  warna, biji, nama, rasa):
        
        #turunan parents
        super().__init__(warna, biji)
        
        #milik child
        self.__nama = nama
        self.__rasa = rasa
        
    #getter
    def getNama(self):
        return self.__nama
    def getRasa(self):
        return self.__rasa
    
    #setter
    def setNama(self, nama) :
        self.__nama = nama
    
    def setRasa(self, rasa) :
        self.__rasa = rasa
    
    def infoMelon(self):
        #memanggil method parents
        super().info()
        #method child
        print(f"warna: {self.__nama} \nrasa: {self.__rasa}")
        
        
#class blueberry inheritence ke class buah        
class Blueberry(Buah) :
    def __init__(self,  warna, biji, nama, rasa):
        
        #turunan parents
        super().__init__(warna, biji)
        
        #milik child
        self.__nama = nama
        self.__rasa = rasa
        
    #getter
    def getNama(self):
        return self.__nama
    def getRasa(self):
        return self.__rasa
    
    #setter
    def setNama(self, nama) :
        self.__nama = nama
    
    def setRasa(self, rasa) :
        self.__rasa = rasa
    
    def infoBlueberry(self):
        #memanggil method parents
        super().info()
        #method child
        print(f"warna: {self.__nama} \nrasa: {self.__rasa}")
        
        
print("Obj info dari class buah")
buah1 = Buah ("orange", True)
buah1.info()

print("\nObj info dari class melon")
melon = Melon ("hijau", True, "melon", "manis")
melon.infoMelon()

print("\nObj info dari class blueberry")
blueberry = Blueberry ("orange", True, "blueberry", "manis")
blueberry.infoBlueberry()

print("\nMethod dari abstract class\n")
buah1.ditanam()

print("\nAplikasi setter getter")
print("\nMerubah nama melon")
melon.setNama("Melon Hijau")
melon.infoMelon()

print("\nMenampilkan rasa blueberry")
blueberry.getRasa()
```

```python
import custModules as cm

#memanggil method di custom modules
cm.judul()

#Abstraction
class Abstract3(ABC):
    @abstractmethod
    def namaOperasi() :
        pass

class OperasiMath(object) :
    def __init__(self, x, y) :
        self.x = x
        self.y = y
        
    def info(self) :
        print(f"nilai x: {self.x} \nnilai y: {self.y}")
        
    def namaOperasi(self) :
        print("penjumlahan,perkalian, pembagian")
        
#class penjumlahan inheritence ke class operasimath      
class Penjumlahan(OperasiMath) :
    def __init__(self, x, y):
        
        #turunan parents
        super().__init__(x, y)
        
        #milik child
        self.__nama = "Penjumlahan"
        
    #getter
    def getNama(self):
        return self.__nama
    
    #setter
    def setNama(self, nama) :
        self.__nama = nama
        
    def inputpenjumlahan(self) :
        #method child
        print(f"{self.__nama}")
        self.x = int(input("masukkan nilai x :"))
        self.y = int(input("masukkan nilai y :"))
        op = cm.penjumlahan(self.x,self.y)
        op.lakukanOperasi()
    
    def penjNormal(self):
        #method child
        print(f"{self.__nama}")
        op = cm.penjumlahan(self.x,self.y)
        op.lakukanOperasi()

#class pembagian inheritence ke class operasimath      
class Pembagian(OperasiMath) :
    def __init__(self, x, y):
        
        #turunan parents
        super().__init__(x, y)
        
        #milik child
        self.__nama = "Pembagian"
        
    #getter
    def getNama(self):
        return self.__nama
    
    #setter
    def setNama(self, nama) :
        self.__nama = nama
    
    def pembNormal(self):
        #method child
        print(f"{self.__nama}")
        op = cm.pembagian(self.x,self.y)
        op.lakukanOperasi()
       
#class perkalian inheritence ke class operasimath      
class Perkalian(OperasiMath) :
    def __init__(self, x, y):
        
        #turunan parents
        super().__init__(x, y)
        
        #milik child
        self.__nama = "Perkalian"
        
    #getter
    def getNama(self):
        return self.__nama
    
    #setter
    def setNama(self, nama) :
        self.__nama = nama
    
    def perkNormal(self):
        #method child
        print(f"{self.__nama}")
        op = cm.perkalian(self.x,self.y)
        op.lakukanOperasi()    

print("PENJUMLAHAN\n\n")
op1 = Penjumlahan(5,3)
print("penjumlahan tanpa inputan\n")
op1.penjNormal()
print("\n\npenjumlahan melalui inputan\n")
op1.inputpenjumlahan()

print("\n\nPEMBAGIAN\n\n")
op2 = Pembagian(6,2)
print("pembagian tanpa inputan\n")
op2.pembNormal()


print("\n\nPERKALIAN\n\n")
op2 = Perkalian(11,3)
print("perkalian tanpa inputan\n")
op2.perkNormal()
```
